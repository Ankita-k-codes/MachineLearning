```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#ploting the data
#sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding titels x axis y axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python


plt.plot(x,y, 'b*')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x), 'g-')
plt.show()
```

```python
#bar plot

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English' , 'Hindi' , 'Sanskrit' , 'Tamil', 'Urdu']
people = [20, 100, 30, 40,10]
ax.bar(languages, people )
plt.xlabel('LANGUAGES')
plt.ylabel('NO. OF PEOPLE')
plt.show()
```

```python
# pie chart
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English' , 'Hindi' , 'Sanskrit' , 'Tamil', 'Urdu']
people = [20, 100, 30, 40,10]
ax.pie(people, labels=languages, autopct = '%1.1f%%')
plt.show()
```

```python
#scatter plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

```python
#3d scatter plot

fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greys')
plt.show()
```
